```python

# Goal => Predict which customer will buy new product lets say (SUV). 
# We know their salary because to purchase lets say a car, you need credit. 

# Import Libraries 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# Importing the dataset 
theDataset = pd.read_csv('Social_Network_Ads.csv')
```

```python
# Check the dataset 
print(theDataset)
```

```python
# Splitting the Dataset into X(dependent) and y(independent) variables
X = theDataset.iloc[:, :-1].values 
y = theDataset.iloc[:, -1].values 
```

```python
print(X)
```

```python
print(y) 
```

```python
# Splitting the dataset into training and testing set 
from sklearn.model_selection import train_test_split
```

```python
# now allocate some parts of the dataset to training and some part to testing 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=101)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

```python
# Feature Scaling?? all features are all on the same scale so each feature is equally important 
# and makes it easier to process machine learning algorithm. 
# Standardisation => features rescaled to ensure the mean and standard deviation to be 0 and 1.
# Redistribution value to be 0 and 1, this is useful for optimisation algos like gradient descent 
# that is used within ml algos that weight inputs(ie regression and neural nets) 
# Rescaling is also used to for algos that use distance measurements ie. KNN
from sklearn.preprocessing import StandardScaler 
standardScaler = StandardScaler() 
```

```python
X_train = standardScaler.fit_transform(X_train) # fit to data, then transform it. 
X_test = standardScaler.transform(X_test) # peform standardisation by centering and scaling 
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
# Training the Logistic Regression Model 
from sklearn.linear_model import LogisticRegression 
# create an instance of the logistic regression class 
classifier = LogisticRegression() 
```

```python
classifier.fit(X_train, y_train)
```

```python
# Predicting a new result 
print(classifier.predict(standardScaler.transform([[30,87000]])))
```

```python
# Predicting the test results 
y_pred = classifier.predict(X_test) 
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

```python
# Making a Confusion Matrix 
# Explanation => table used to describe the performance of a classification model. 
# True Positive => Predicted value Yes is the same as actual value Yes. 
# False Positive => Predicted value Yes is not the same as actual value No. 
# False Negative => Predicted Value No is not the same as actual value Yes. 
# True Negative => Predicted value No is the same as actual value No. 

# Model has 82% accuracy. 

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm) 
accuracy_score(y_test, y_pred)
```

```python
# Now Visualise the training set 
# Either use matplit lib or export training set(both X and y) which is inverse transformed and 
# plot them in tableau. 

# Using matplotlib => read this code later. 
from matplotlib.colors import ListedColormap
X_set, y_set = standardScaler.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(standardScaler.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
# visualise the test set using matplotlib 
from matplotlib.colors import ListedColormap
X_set, y_set = standardScaler.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(standardScaler.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
